#  Behind the Numbers: Gross Profit Trends in Tech


```python
# importing required libraries

import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

# Brief Description of Technology Companies and Their Gross Profits

## Companies Overview

We will be comparing the gross profits of five prominent technology companies: Twitter, Snap, Pinterest, Match Group, and Zoom. Each of these companies operates in various sectors of the technology industry and has gained significant attention in recent years.

## Company Descriptions

1. **Twitter**: A well-known social media platform that enables users to post and interact with short messages known as "tweets."

2. **Snap**: The parent company of Snapchat, a multimedia messaging app known for its disappearing messages and various multimedia features.

3. **Pinterest**: A visual discovery and bookmarking platform that allows users to find and save ideas for various projects and interests.

4. **Match Group**: The company behind several popular online dating platforms, including Tinder, Match.com, and OkCupid.

5. **Zoom**: A video communication platform that gained immense popularity for its virtual meetings and web conferencing capabilities.

These companies have diverse focuses within the technology sector, and analyzing their gross profits can provide insights into their financial performance and market presence.

## Gross Profits Table

Here is a table presenting the gross profits of these companies over the past 5 years (2018-2022):

| Company      | 2018   | 2019   | 2020   | 2021   | 2022   |
|--------------|--------|--------|--------|--------|--------|
| Twitter      | 2080M | 2320M | 2350M | 3280M | 3180M |
| Snap         | 382M  | 820M  | 1324M | 2367M | 2787M |
| Pinterest    | 514M  | 784M  | 1243M | 2049M | 2124M |
| Match Group  | 1320M | 1524M | 1755M | 2144M | 2229M |
| Zoom         | 121M  | 270M  | 507M  | 1829M | 3045M |

These figures represent the gross profits (in millions($)) of each company for the years 2018 to 2022.


```python
companies = ['Twitter', 'Snap', 'Pinterest', 'Match Group', 'Zoom']
colors = ['blue', 'orange', 'green', 'red', 'purple']  # Add more colors if needed

# dictionary created with each company's gross profit

twitter = {"2018":2080, "2019": 2320, "2020":2350, "2021":3280, "2022":3180}
snap = {"2018": 382, "2019":820, "2020":1324, "2021":2367, "2022":2787}
pinterest ={"2018":514, "2019":784, "2020":1243, "2021":2049, "2022":2124}
match_group= {"2018":1320, "2019":1524, "2020":1755, "2021":2144, "2022":2229}
zoom = {"2018":121, "2019":270, "2020":507, "2021":1829, "2022":3045}
```

## Line Charts Representing Gross Profit for Each Company

```python
data = [twitter, snap, pinterest, match_group, zoom]

# Creating individual charts for each company's total profit
for i, company in enumerate(companies):
    plt.figure(figsize=(8, 6))
    years = list(data[i].keys())
    profits = list(data[i].values())
    plt.plot(years, profits, marker='o', color=colors[i])
    plt.title(f'Gross Profit (2018-2022) for {company}')
    plt.xlabel('Year')
    plt.ylabel('Gross Profit (in Millions ($))')
    plt.grid(True)
    plt.show()
```

## Combined Chart for 5 Companies 

```python
# Combined chart for each company's total profit
plt.figure(figsize=(10, 6))
line_styles = ['-', '--', '-.', ':', '-']  # Different line styles for each company
for i, (data, label) in enumerate(zip([twitter, snap, pinterest, match_group, zoom], companies)):
    years = list(data.keys())
    profits = list(data.values())
    plt.plot(years, profits, linestyle=line_styles[i], marker='o', label=label)

plt.title('Gross Profit Comparisons')
plt.xlabel('Year')
plt.ylabel('Gross Profit (in Millions ($))')
plt.legend()
plt.grid(True)
plt.show()
```

## Network Graph for Each Company

```python
# Data
companies = ['Twitter', 'Snap', 'Pinterest', 'Match Group', 'Zoom']
profits = [twitter, snap, pinterest, match_group, zoom]
node_colors = ['blue', 'orange', 'green', 'red', 'purple']


# Creating a network graph for each company with different colors
for i, (company, profit_data) in enumerate(zip(companies, profits)):
    G = nx.Graph()
    
    plt.figure(figsize=(10, 6))
    plt.title(f'Network Graph for {company}')

    for year, profit in profit_data.items():
        G.add_node(year, bipartite=0)
        G.add_edge(company, year, weight=profit)
        
    edge_labels = {(u, v): f"${data}M" for u, v, data in G.edges(data='weight')}
    
    pos = nx.spring_layout(G, seed=42)
    
    nx.draw_networkx_nodes(G, pos, nodelist=[company], node_color=node_colors[i], node_size=2000)
    nx.draw_networkx_nodes(G, pos, nodelist=list(profit_data.keys()), node_color='lightblue', node_size=1000)
    nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')
    nx.draw_networkx_edges(G, pos)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, bbox=dict(boxstyle="round,pad=0.2", edgecolor="gray", facecolor="white"))

    plt.xticks([])
    plt.yticks([])
    plt.show()
```
